```python
# Tanh Activation Function & Derivative Example - Sentiment Analysis
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Generate synthetic sentiment "raw scores" (before activation)
# Negative reviews (centered at -2), Positive reviews (centered at +2)
np.random.seed(42)
negative_reviews = np.random.normal(loc=-2, scale=1.0, size=50)  # mostly negative
positive_reviews = np.random.normal(loc=2, scale=1.0, size=50)   # mostly positive

# Combine into one dataset
z_values = np.concatenate([negative_reviews, positive_reviews])
```

```python
# 2. Define tanh and its derivative
def tanh(z):
    return np.tanh(z)

def tanh_derivative(z):
    return 1 - np.tanh(z)**2
```

```python

# 3. Compute outputs
tanh_outputs = tanh(z_values)
tanh_derivatives = tanh_derivative(z_values)
```

```python
# 4. Plot tanh activation and derivative
z_range = np.linspace(-5, 5, 400)
plt.figure(figsize=(9, 5))

# Plot activation curve
plt.plot(z_range, tanh(z_range), label="tanh(z) → Sentiment Output", color="blue", linewidth=2)

# Plot derivative curve
plt.plot(z_range, tanh_derivative(z_range), label="tanh'(z) → Learning Speed", color="orange", linestyle="--", linewidth=2)

# Plot synthetic data points
plt.scatter(z_values, tanh_outputs, color="blue", alpha=0.5, label="Synthetic Sentiment Data")

# Highlight zones
plt.fill_between(z_range, tanh(z_range), 0, where=(tanh(z_range) > 0), color="blue", alpha=0.05)
plt.fill_between(z_range, tanh(z_range), 0, where=(tanh(z_range) < 0), color="red", alpha=0.05)

# Lines at center
plt.axhline(0, color="black", linewidth=0.7, linestyle="--")
plt.axvline(0, color="gray", linewidth=0.7, linestyle="--")

# Labels & title
plt.title("Tanh Activation in Sentiment Analysis with Derivative")
plt.xlabel("z (Raw Sentiment Score)")
plt.ylabel("Output / Derivative Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Generate predict function
def predict_sentiment(tanh_output):
    """
    Predicts sentiment based on tanh output.

    Args:
        tanh_output: The output from the tanh activation function.

    Returns:
        A string indicating the predicted sentiment ("Positive", "Negative", or "Neutral").
    """
    if tanh_output > 0.1:  # Threshold for positive sentiment
        return "Positive"
    elif tanh_output < -0.1: # Threshold for negative sentiment
        return "Negative"
    else:
        return "Neutral"

# Example usage with synthetic data points
predicted_sentiments = [predict_sentiment(output) for output in tanh_outputs]

# Display some examples
for i in range(10):
    print(f"Raw score: {z_values[i]:.2f}, Tanh output: {tanh_outputs[i]:.2f}, Predicted sentiment: {predicted_sentiments[i]}")
```
