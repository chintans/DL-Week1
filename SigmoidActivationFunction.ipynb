```python
# Sigmoid Activation Function Example - Spam Detection (Binary Classification)
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Generate synthetic "spam likelihood" data
# Feature: suspicious score (0 = safe, 10 = very suspicious)
np.random.seed(42)
safe_emails = np.random.normal(loc=2, scale=1.0, size=50)  # mostly safe
spam_emails = np.random.normal(loc=8, scale=1.0, size=50)  # mostly spam

X = np.concatenate([safe_emails, spam_emails])
y = np.array([0]*50 + [1]*50)  # 0 = Not Spam, 1 = Spam
```

```python
# 2. Define sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
# 3. Assume a simple model: z = w*x + b
w = 1.2   # weight
b = -6    # bias
z_values = w * X + b
sigmoid_outputs = sigmoid(z_values)
```

```python
# 4. Plot results
plt.figure(figsize=(8,5))

# Plot raw data points
plt.scatter(X, y, color="blue", label="Actual Labels", alpha=0.5)

# Plot sigmoid curve
x_range = np.linspace(0, 10, 200)
z_range = w * x_range + b
plt.plot(x_range, sigmoid(z_range), color="red", linewidth=2, label="Sigmoid Output (Predicted Probability)")

# Mark threshold line at 0.5
plt.axhline(0.5, color="gray", linestyle="--")
plt.text(10.2, 0.48, "Decision Threshold (0.5)", fontsize=9, va="center")

plt.title("Sigmoid Activation for Spam Detection (Synthetic Data)")
plt.xlabel("Suspicious Score (0 = Safe, 10 = Very Suspicious)")
plt.ylabel("Predicted Probability of Spam")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 5. Define a prediction function
def predict(x, w, b, threshold=0.5):
    """
    Predicts the class (0 or 1) based on the sigmoid output and a threshold.
    """
    z = w * x + b
    sigmoid_output = sigmoid(z)
    return 1 if sigmoid_output >= threshold else 0

# Example predictions
example_suspicious_score_1 = 3
example_suspicious_score_2 = 7

prediction_1 = predict(example_suspicious_score_1, w, b)
prediction_2 = predict(example_suspicious_score_2, w, b)

print(f"For a suspicious score of {example_suspicious_score_1}, the prediction is: {prediction_1}")
print(f"For a suspicious score of {example_suspicious_score_2}, the prediction is: {prediction_2}")
```
